```python
import pandas as pd
import numpy as np
import scipy as sp
def logloss(act, pred):
    epsilon = 1e-15
    pred = sp.maximum(epsilon, pred)
    pred = sp.minimum(1-epsilon, pred)
    ll = np.sum(act*sp.log(pred) + sp.subtract(1,act)*sp.log(sp.subtract(1,pred)))
    ll = ll * -1.0/len(act)
    return ll

print logloss([1,1],[0.1,0.4])

train_data = pd.read_csv('train.csv').sample(5000, random_state=23)
test_data = pd.read_csv('test.csv').sample(5000, random_state=23).sort_values("test_id")
```

```python
import re, time
from nltk.corpus import stopwords
from nltk import WordNetLemmatizer, word_tokenize
from nltk.stem.porter import PorterStemmer

def word_extractor(text):
    stemmer = PorterStemmer()
    commonwords = stopwords.words('english')
    try:
        text = re.sub(r'([a-z])\1+', r'\1\1',text)#substitute multiple letter by two
        text = re.sub(r'([^\s\w]|_)+', r'',text) #delete chars that are not letter or number and downcase
        text = re.sub(r'[^\w]', r' ',text)
    except: 
        if text != text:
            text = " "
        print text
    words = []
    wordtokens = [ stemmer.stem(word.lower())  for word in word_tokenize(text.decode('utf-8', 'ignore')) ]
    for word in wordtokens:
        if word not in commonwords:
            words.append(word)
    return words

question1 =  train_data.iloc[0][3]
question2 = train_data.iloc[0][4]
print question1
print question2
print 1==train_data.iloc[0].is_duplicate

print word_extractor(question1)
print word_extractor(question2)

#tokeniza
pregunta1 = [word_extractor(word) for word in train_data.question1 ]
pregunta2 = [word_extractor(word) for word in train_data.question2 ]

pregunta1_test = [word_extractor(word) for word in test_data.question1 ]
pregunta2_test = [word_extractor(word) for word in test_data.question2 ]
print "listo"
```

```python
from sklearn.metrics import mutual_info_score as MI
from sklearn.metrics import normalized_mutual_info_score as NMI
from sklearn.metrics import fowlkes_mallows_score as FMS
from sklearn.metrics import jaccard_similarity_score as JS
from sklearn.metrics import v_measure_score as VMS
from sklearn import cross_validation

def C_V(model, Xm,ym, K=10):
    try:
        k_fold = cross_validation.KFold(len(Xm),K)
    except:
        k_fold = cross_validation.KFold(Xm.shape[0],K)
    error_cv = 0
    for k, (train, val) in enumerate(k_fold):
        model.fit(Xm[train], ym[train])
        yhat_val = model.predict(Xm[val])
        mse_fold = logloss(ym[val],model.predict_proba(Xm[val])[:,1])
        #mse_fold = np.mean(np.power(yhat_val - ym[val], 2))
        error_cv += mse_fold
    error_cv = error_cv / K
    return error_cv

def precict_similarity(p1,p2):
    informacion = []
    vectorizer = TfidfVectorizer(ngram_range=(1, 1), binary='False')
    vectorizer.fit(np.asarray(p1+p2)) #para normal mutual imfromation
    
    for i in range(5000):
        q1 = [p1[i]] #string de la pregunta 1
        q2 = [p2[i]]

        try:
            #vectorizer.fit(np.asarray(q1+q2))
            aux1= vectorizer.transform(q1).todense().tolist()[0]
            aux2= vectorizer.transform(q2).todense().tolist()[0]
            #medida de similaridad
            #informacion+= [NMI(aux1,aux2)]
            #informacion+= [FMS(aux1,aux2)]
            #informacion+= [JS(aux1,aux2)]
            informacion+= [VMS(aux1,aux2)]
        except:
            informacion+= [0] #NAN
            pass
    return informacion

ytrain = np.asarray(train_data.is_duplicate)

#Mutual information
#ypred = precict_similarity(pregunta1,pregunta2)
#print "LogLoss en Mutual Information predict model: %f"%logloss(ytrain,ypred)

#train_data["Similaridad"] = ypred


xtrain = crear_representacion(pregunta1,pregunta2)

from sklearn.model_selection import cross_val_score as CV

from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
model = model.fit(xtrain, ytrain)
print "Naive Bayes"
print "LogLoss en Train set predict model: %f"%logloss(ytrain,model.predict_proba(xtrain)[:,1])
print "Cross validation predict model: %f"%C_V(model,xtrain,ytrain,10)
#score_the_model(model,x,y,xt,yt,"BernoulliNB")

from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model = model.fit(xtrain,ytrain)
print "Multinomial Bayes"
print "LogLoss en Train NB predict model: %f"%logloss(ytrain,model.predict_proba(xtrain)[:,1])
print "Cross validation predict model: %f"%C_V(model,xtrain,ytrain,10)

from sklearn.ensemble import RandomForestClassifier as RandomFor
model = RandomFor(max_depth=15)
model = model.fit(xtrain,ytrain)
print "Random Forest"
print "LogLoss en Train predict model: %f"%logloss(ytrain,model.predict_proba(xtrain)[:,1])
print "Cross validation predict model: %f"%C_V(model,xtrain,ytrain,10)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(penalty='l2',C=0.1)
model = model.fit(xtrain,ytrain)
print "Logistic Regression"
print "LogLoss en Train  predict model: %f"%logloss(ytrain,model.predict_proba(xtrain)[:,1])
print "Cross validation predict model: %f"%C_V(model,xtrain,ytrain,10)

from sklearn.ensemble import AdaBoostClassifier as ADA
model = ADA()
model = model.fit(xtrain,ytrain)
print "ADA"
print "LogLoss en Train  predict model: %f"%logloss(ytrain,model.predict_proba(xtrain)[:,1])
print "Cross validation predict model: %f"%C_V(model,xtrain,ytrain,10)
```

```python
def prob_format(p):
    #return '%e' % p
    return round(p,1)

def write_submit(ids, probs, file_name):
    assert len(ids) == len(probs)
    with open(file_name, 'w') as f:
        f.write('test_id,is_duplicate\n')
        for i, p in zip(ids, probs):
            f.write('%s,%s\n' % (i, prob_format(p)))
    print('*wrote %s' % file_name)

xtest =  crear_representacion(pregunta1_test,pregunta2_test)
print "vector de representacion creada"
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(penalty='l2',C=0.1)
model = model.fit(xtrain,ytrain)
print "modelo ajustado"
write_submit(np.asarray(test_data.test_id),model.predict_proba(xtest)[:,1],"MiprimeraEntrega.csv")
```
