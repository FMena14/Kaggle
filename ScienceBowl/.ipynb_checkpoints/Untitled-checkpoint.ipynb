```python
import dicom
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


from subprocess import check_output

etiqueta = pd.read_csv('stage1_labels.csv')
def buscar_etiqueta(patient_id):
    return int(etiqueta.loc[etiqueta.id == patient_id].cancer)
        

carpetas = check_output(["ls", "sample_images/"]).decode("utf8")


X_train = []
etiquetas = []
ids = []
for i in range(2,5):
    print "esta leyendo la carpeta "+str(i)

    #lee carpeta
    primera_carpeta = carpetas.split('\n')[i]

    #lee subcarpeta
    archivos = check_output(["ls", "sample_images/"+primera_carpeta]).decode("utf8")

    archivos_lista = archivos.split('\n')

    archivos_dicom = []
    for i in archivos_lista:
        if len(i)!=0:
            archivos_dicom += [dicom.read_file('sample_images/'+primera_carpeta+'/'+i)]

    #para cada imagen del paciente:
    for imagen in archivos_dicom:
        slice = imagen.pixel_array       #sacar imagen/matriz
        slice[slice == -2000] = 0
        
        #pasar a vector
        vector_imagen = np.concatenate(slice)
        
        #obtener id
        ids.append(imagen.PatientID)
        etiquetas.append( buscar_etiqueta(imagen.PatientID) )
                
        #matriz de entrenamiento
        X_train.append(vector_imagen)
    

X_train = pd.DataFrame.from_records(X_train)
print "termino"
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation,Dropout
from keras.optimizers import SGD,RMSprop

model = Sequential()

model.add(Dense(100, input_dim=X_train.shape[1], init='uniform', activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1, init='uniform', activation='sigmoid'))
model.compile(optimizer=SGD(lr=0.05), loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train.as_matrix(), np.asarray(etiquetas), nb_epoch=50, batch_size=32, verbose=1)

```

```python
scores = model.evaluate(X_train.as_matrix(), etiquetas)
print "\nVALIDATION ACCURACY (activation relu-sigmoid)= %f"% scores[1]
```
